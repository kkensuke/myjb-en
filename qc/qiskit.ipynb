## Basics

### Make a quantum circuit

`QuantumCircuit` is a class to make a quantum circuit. The argument `1` means that the quantum circuit has one qubit. The qubits indeces are numbered from 0.

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)
```

### Add a quantum gate

Generated quantum circuit is empty. We add a quantum gate to the quantum circuit to generate a quantum state: $\frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$.

```python
qc.h(0)
```

### Draw a quantum circuit

```python
qc.draw(output='mpl')
```

### Show the state vector on the bloch sphere

```python
from qiskit.visualization import plot_bloch_multivector

plot_bloch_multivector(qc)
```

### Add a measurement

To extract information from a quantum circuit, we need to perform measurements.

```python
qc.measure_all()
```

### Simulate the quantum circuit

```python
from qiskit import Aer, execute

backend = Aer.get_backend('qasm_simulator')
result = execute(qc, backend).result()
counts = result.get_counts(qc)
print(counts)
```

### Visualize the result

```python
from qiskit.visualization import plot_histogram

plot_histogram(counts)
```
